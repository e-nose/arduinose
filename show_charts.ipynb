```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "plotly_mimetype"

columns_default = ['RX Date/Time', 
                  'MQ2 (Methane, Butane, LPG)', 
                  'MQ3 (Alcohol, Ethanol)', 
                  'MQ4 (Methane, CNG Gas)', 
                  'MQ7 (Carbon Monoxide)', 
                  'MQ8 (Hydrogen Gas)', 
                  'MQ9 (Carbon Monoxide, flammable gases)',
                  'MQ135 (CO, Ammonia, Benzene, Alcohol)'
                  ]

color_discrete_map = {
                  'MQ2 (Methane, Butane, LPG)':'blue', 
                  'MQ3 (Alcohol, Ethanol)':'orange', 
                  'MQ4 (Methane, CNG Gas)':'mediumseagreen', 
                  'MQ7 (Carbon Monoxide)':'red', 
                  'MQ8 (Hydrogen Gas)':'mediumpurple', 
                  'MQ9 (Carbon Monoxide, flammable gases)':'sienna',
                  'MQ135 (CO, Ammonia, Benzene, Alcohol)':'violet'
}

def show_chart(filename):
    path = 'csv/' + filename + '.csv'
    df = pd.read_csv(path)
#     print(len(df))
#     df = df.head(300)
    df = df.tail(600)
    df.columns = columns_default
    fig = px.line(df, x = "RX Date/Time", y = df.columns, title = filename, color_discrete_map = color_discrete_map)

    fig.update_xaxes(showticklabels = False)
    fig.show()
#     print('datapoints recorded:',len(df))
#     display(df.max())
#     display(df.min())
    print('------------------------------------------------------------------------------------------------------------')
```
