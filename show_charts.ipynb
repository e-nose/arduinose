```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "plotly_mimetype"

def show_chart(filename):
    path = 'csv/' + filename + '.csv'
    df = pd.read_csv(path)
#     print(len(df))
#     df = df.head(300)
#     df = df.tail(250)
    fig = px.line(df, x = "RX Date/Time", y = df.columns, title=filename)

    fig.update_xaxes(showticklabels = False)
    fig.show()
    print('datapoints recorded:',len(df))
    print('------------------------------------------------------------------------------------------------------------')
```

```python
show_chart('sanitiser')
# show_chart('10g_semi-fresh_coffee')
show_chart('10g_fresh_cheap')
show_chart('less_coffee')
show_chart('03_25_17-58_old_ground_2_pumps')
```
