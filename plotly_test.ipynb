```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "plotly_mimetype"

pc = pd.read_csv('csv/pc.csv')
vin = pd.read_csv('csv/vin.csv')
display(pc.head())
display(vin.head())
```

```python
pc.columns = ['RX Date/Time', 'MQ2(this)', 'MQ3(field 2)', 'MQ4(field 3)',
       'MQ6(field 4)', 'MQ7(field 5)', 'MQ8(field 6)', 'MQ9(field 7)',
       'MQ135(field 8)']
pc
```

```python
display(pc.describe())
print(pc['MQ3(field 2)'].max())
```

```python
display(vin.describe())
print(vin['MQ3(field 2)'].max())
```

```python
px.line(pc, x="RX Date/Time", y="MQ4(field 3)")
```

```python
fig1 = px.line(pc, x="RX Date/Time", y=pc.columns)
fig2 = px.line(vin, x="RX Date/Time", y=vin.columns)

fig1.update_xaxes(showticklabels=False)
fig1.show()
fig2.update_xaxes(showticklabels=False)
fig2.show()
```

```python
vin_test = vin.copy()
vin_above_mean = vin_test[vin_test['MQ2(field 1)'] > vin_test['MQ2(field 1)'].mean()]
display(vin_above_mean.head())
print(len(vin_above_mean), 'datapoints are above mean regarding MQ2 (out of', len(vin_test), ')')
```

```python
fig3 = px.line(vin_above_mean, x="RX Date/Time", y=vin.columns)
fig3.update_xaxes(showticklabels=False)

fig2.show()
fig3.show()
```
